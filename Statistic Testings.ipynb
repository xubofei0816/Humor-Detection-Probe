```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
```

```python
type1_non_punchs = pd.read_csv('type1_non_punchs.csv')
type1_punchs = pd.read_csv('type1_punchs.csv')
type0_non_punchs = pd.read_csv('type0_non_punchs.csv')
type0_punchs = pd.read_csv('type0_punchs.csv')
type0_set_ups = pd.read_csv('type0_set_ups.csv')
```

```python
print(type1_non_punchs.shape)
print(type1_punchs.shape)
```

```python
array1 = type1_punchs['0'].to_numpy()
array2 = type1_non_punchs['0'].to_numpy()
array3 = type0_punchs['0'].to_numpy()
array4 = type0_non_punchs['0'].to_numpy()
array5 = type0_set_ups['0'].to_numpy()
```

```python
# make a new arry with all data
# append arrays together
array = np.append(array1, array2)
array = np.append(array, array3)
array = np.append(array, array4)
array_all = np.append(array, array5)
# make a new array of 0 and 1, make all data from array2, array4, array5 as 1, and all data from array1, array3 as 0
array_all_label = np.vstack((np.zeros((array1.shape[0], 1)), np.ones((array2.shape[0], 1)), np.zeros((array3.shape[0], 1)), np.ones((array4.shape[0], 1)), np.ones((array5.shape[0], 1))))
array_all_label = array_all_label.reshape(-1,)

```

```python

# Perform Pearson correlation coefficient
corr_coefficient, p_value = pearsonr(array_all, array_all_label)

# Interpret the results
print(f'Pearson Correlation Coefficient: {corr_coefficient}')
print(f'P-value: {p_value}')

# Interpretation
if p_value < 0.05:
    print('There is a significant correlation between the two arrays.')
else:
    print('There is no significant correlation between the two arrays.')
```

```python
from scipy.stats import pointbiserialr

# Perform Point-Biserial correlation coefficient
corr_coefficient, p_value = pointbiserialr(array_all, array_all_label)

# Interpret the results
print(f'Point-Biserial Correlation Coefficient: {corr_coefficient}')
print(f'P-value: {p_value}')

# Interpretation
if p_value < 0.05:
    print('There is a significant correlation between the two arrays.')
else:
    print('There is no significant correlation between the two arrays.')
```

```python
from scipy.stats import mannwhitneyu


# Perform Mann-Whitney U test for each pair of arrays
arrays = [array1, array2, array3, array4, array5]

# Store results in a DataFrame
results = pd.DataFrame(columns=['Arrays', 'Statistic', 'P-value'])
```

```python
significance_level = 0.05

# Interpret the results based on the p-values
for i in range(len(arrays)):
    for j in range(i+1, len(arrays)):
        stat, p_value = mannwhitneyu(arrays[i], arrays[j])
        #results = results.append({'Arrays': f'{i+1} vs {j+1}', 'Statistic': stat, 'P-value': p_value}, ignore_index=True)
        #results = pd.concat(results, {'Arrays': f'{i+1} vs {j+1}', 'Statistic': stat, 'P-value': p_value}, ignore_index=True)
        results = pd.concat([results, pd.DataFrame({'Arrays': [f'{i+1} vs {j+1}'], 'Statistic': [stat], 'P-value': [p_value]})], ignore_index=True)

        if p_value < significance_level:
            print(f'Reject the null hypothesis for Arrays {i+1} and {j+1}:')
            print(f'  Statistic={stat}, p-value={p_value}')
            print('  There is evidence of a significant difference in the distributions.')
        else:
            print(f'Fail to reject the null hypothesis for Arrays {i+1} and {j+1}:')
            print(f'  Statistic={stat}, p-value={p_value}')
            print('  There is not enough evidence to conclude a significant difference in the distributions.')

# Set the index to 'Arrays'
results.set_index('Arrays', inplace=True)

results['Significant'] = results['P-value'] < significance_level

# Display the results
print(results)
```
