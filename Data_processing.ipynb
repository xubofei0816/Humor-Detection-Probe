```python
# Function to find the indices of Punch_tokens in Joke_tokens
def find_indices(row):
    joke_tokens = list(row['Joke_tokens'].split())
    punch_tokens = list(row['punch_tokens'].split())

    # Initialize list to store indices
    indices = []

    # Iterate over punch_tokens and find indices in joke_tokens
    for punch_token in punch_tokens:
        try:
            punch_indices = [i for i, token in enumerate(joke_tokens) if token == punch_token]
            indices.extend(punch_indices)
        except ValueError:
            # Handle the case where the token is not found
            indices.append(None)

    return indices[1:-1]

# Apply the function to each row in the DataFrame
jokes1['punch_token_index'] = jokes1.apply(find_indices, axis=1)
```

```python
def filter_consecutive(lst):
    lst.sort()
    result = []
    for i in range(len(lst)):
        if i == 0 or lst[i] == lst[i - 1] + 1:
            result.append(lst[i])
    return result
```

```python
# Create a function to format each row
def format_row_tr(row):
    return 'tr'+f"\t{row['Type']}\t{row['Joke']}"

def format_row_va(row):
    return 'va'+ f"\t{row['Type']}\t{row['Joke']}"

def format_row_te(row):
    return 'te' + f"\t{row['Type']}\t{row['Joke']}"
```

```python
def split_data(df, train_prop=0.7, val_prop=0.15, test_prop=0.15, random_seed=None):
    """
    Split a DataFrame into training, validation, and test sets.

    Parameters:
    - df: DataFrame
    - train_prop: float, proportion of data for training (default is 0.7)
    - val_prop: float, proportion of data for validation (default is 0.15)
    - test_prop: float, proportion of data for testing (default is 0.15)
    - random_seed: int, seed for reproducibility (default is None)

    Returns:
    - train_df, val_df, test_df: DataFrames representing training, validation, and test sets
    """

    # Check if the proportions sum to 1.0
    assert train_prop + val_prop + test_prop == 1.0, "Proportions must sum to 1.0"

    # Set random seed for reproducibility
    np.random.seed(random_seed)

    # Shuffle the DataFrame
    shuffled_df = df.sample(frac=1, random_state=random_seed).reset_index(drop=True)

    # Calculate the number of rows for each set
    num_rows = len(df)
    train_size = int(train_prop * num_rows)
    val_size = int(val_prop * num_rows)

    # Split the DataFrame
    train_df = shuffled_df[:train_size]
    val_df = shuffled_df[train_size:train_size + val_size]
    test_df = shuffled_df[train_size + val_size:]

    return train_df, val_df, test_df
```

```python
train_df, val_df, test_df = split_data(jokes, train_prop=0.7, val_prop=0.15, test_prop=0.15, random_seed=42)
# Apply the formatting functions to each set
formatted_train = train_df.apply(format_row_tr, axis=1)
formatted_val = val_df.apply(format_row_va, axis=1)
formatted_test = test_df.apply(format_row_te, axis=1)

# Combine the formatted sets into one DataFrame
formatted_combined = pd.concat([formatted_train, formatted_val, formatted_test])

# Save the formatted rows to a text file
formatted_combined.to_csv('conflict_type.txt', index=False, header=False)
```

```python
# Open the file in read mode
with open('conflict_type.txt', 'r') as file:
    # Read the contents of the file
    lines = file.readlines()

# Remove quotation marks from the beginning and end of each line
modified_lines = [line.strip('"') for line in lines]

# Open the file in write mode to save the modified content
with open('conflict_type.txt', 'w') as file:
    # Write the modified content back to the file
    file.writelines((modified_lines))

print("File 'conflict_type.txt' created successfully.")
```
